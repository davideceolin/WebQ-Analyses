```python
import xml.etree.ElementTree as ET
from collections import Counter
import os.path
import pandas as pd

vaccinations = pd.read_csv('/Users/dceolin/Downloads/vaccinations_meta.csv',sep=";")
ass_j = pd.read_csv('/Users/dceolin/WebQ-Analyses/data2.csv',sep=";")
path = '/Users/dceolin/Downloads/output 2'

cols = ['Index','overallQuality','first','second','count_cb','count_rob','count_na','count_ncb']
def count_beliefs(x,dir_name,files):
    df = pd.DataFrame([],columns = cols)
    for p in files[1::]:
        f = open(dir_name+"/"+p, 'rt').read()
        root = tree = ET.fromstring("<?xml version=\"1.0\"?><data>"+f+"</data>")
        c = Counter([child.tag for child in root])
        data = [p.split("#")[0],ass_j.loc[ass_j['target']==vaccinations.iloc[int(p.split("#")[0])]['URL']]['neutrality'].mean() ,
                float(c['ROB']+c['NA']+c['NCB'])/c['CB']*100, float(c['ROB'])/c['CB']*100, c['CB'], c['ROB'], c['NA'],c['NCB']]
        df1 = pd.DataFrame([data],columns=cols)
        #print df1
        df = df.append(df1)
    print df.dropna().corr(method='spearman')
        
os.path.walk(path, count_beliefs, 0)
```
